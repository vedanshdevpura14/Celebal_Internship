# Feauture Engineering 

```python
df = df.copy()  

# Ratio features 
df['Production_Efficiency'] = df['Estimated_Deliveries'] / df['Production_Units']  
df['Price_per_km']          = df['Avg_Price_USD'] / df['Range_km']                 
df['CO2_per_delivery']      = df['CO2_Saved_tons'] / df['Estimated_Deliveries']   
```

```python
# Time-based features 
df['Quarter']          = ((df['Month'] - 1) // 3) + 1       # Q1, Q2, Q3, Q4
df['Is_Q4']            = (df['Quarter'] == 4).astype(int)   # Q4 mein Tesla usually zyada deliver karta hai
df['Year_Since_Start'] = df['Year'] - 2015                  # 2015 se ab tak kitne saal
```

```python
# Cyclical encoding 
df['Month_sin'] = np.sin(2 * np.pi * df['Month'] / 12)
df['Month_cos'] = np.cos(2 * np.pi * df['Month'] / 12)
```

```python
# Categorical columns ko numbers mein convert karo — ML model strings nahi samajhta
le_model  = LabelEncoder()
le_region = LabelEncoder()
df['Model_enc']  = le_model.fit_transform(df['Model'])    # Model S=0, X=1....
df['Region_enc'] = le_region.fit_transform(df['Region'])  # Asia=0, Europe=1...
```

```python
new_feats = ['Production_Efficiency','Price_per_km','CO2_per_delivery',
             'Quarter','Is_Q4','Year_Since_Start','Month_sin','Month_cos',
             'Model_enc','Region_enc']
print('Summary:')
print(df[new_feats].describe().round(3))
```

```python
# Train Test Split 

FEATURES = [
    'Year_Since_Start', 'Month_sin', 'Month_cos', 'Quarter', 'Is_Q4',
    'Estimated_Deliveries', 'Production_Units', 'Battery_Capacity_kWh',
    'Range_km', 'Charging_Stations', 'Production_Efficiency',
    'Price_per_km', 'CO2_per_delivery', 'Model_enc', 'Region_enc'
]
TARGET = 'Avg_Price_USD' 

X = df[FEATURES]  
y = df[TARGET]    
# 80% data training ke liye, 20% testing ke liye
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# StandardScaler for normalizing values
scaler    = StandardScaler()
X_train_s = scaler.fit_transform(X_train)  
X_test_s  = scaler.transform(X_test)       # Test pe sirf transform fit nhi krege data leakage hoga

print(f'Train size : {X_train.shape[0]} samples')
print(f'Test size  : {X_test.shape[0]} samples')
print(f'Features   : {len(FEATURES)}')
```

# Training Multiple Regression Models

```python
models = {
    'Linear Regression' : LinearRegression(),                                          
    'Ridge'             : Ridge(alpha=1.0),                                             
    'Lasso'             : Lasso(alpha=0.1),                                            
    'Random Forest'     : RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1), 
    'Gradient Boosting' : GradientBoostingRegressor(n_estimators=100, random_state=42), 
}

results = {}
for name, model in models.items():
    model.fit(X_train_s, y_train)      
    preds = model.predict(X_test_s)    
    results[name] = {
        'R2'  : round(r2_score(y_test, preds), 4),                            
        'MAE' : round(mean_absolute_error(y_test, preds), 2),                
        'RMSE': round(np.sqrt(mean_squared_error(y_test, preds)), 2),        
    }

results_df = pd.DataFrame(results).T.sort_values('R2', ascending=False)
print(' Model Comparison — Top one is best')
print(results_df.to_string())
```

```python
# Model comparision chart 

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
metrics = ['R2', 'MAE', 'RMSE']
colors  = sns.color_palette('Blues_r', len(results_df))

for i, metric in enumerate(metrics):
    data = results_df[metric].sort_values(ascending=(metric != 'R2'))
    bars = axes[i].barh(data.index, data.values, color=colors)
    axes[i].set_title(metric, fontweight='bold')
    for bar, val in zip(bars, data.values):
        axes[i].text(val * 1.01, bar.get_y() + bar.get_height() / 2,
                     f'{val:,.2f}', va='center', fontsize=9)

plt.suptitle('Regression Model Comparison — Predicting Avg Price', fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

# Hyperparameter Tuning Grid searchCV

```python
# Best model we get is random forest now improving it more
param_grid = {
    'n_estimators'     : [50, 100, 200],  
    'max_depth'        : [None, 10, 20],   
    'min_samples_split': [2, 5],           
}
# Total combinations: 3 × 3 × 2 = 18 combinations, har ek 3-fold CV ke saath
# Matlab 18 × 3 = 54 models train honge.

grid_search = GridSearchCV(
    RandomForestRegressor(random_state=42, n_jobs=-1),
    param_grid,
    cv=3,           # 3-fold cross validation
    scoring='r2',   # R2 ke basis pe best model choose karo
    verbose=1,      
    n_jobs=-1      
)
grid_search.fit(X_train_s, y_train)

print('Best Parameters :', grid_search.best_params_)
print('Best CV R2      :', round(grid_search.best_score_, 4))


best_rf    = grid_search.best_estimator_
best_preds = best_rf.predict(X_test_s)

print('\n Tuned RF on Test Set')
print(f'R2   : {r2_score(y_test, best_preds):.4f}')
print(f'MAE  : {mean_absolute_error(y_test, best_preds):,.2f}')
print(f'RMSE : {np.sqrt(mean_squared_error(y_test, best_preds)):,.2f}')
```

```python
# Feature Imortance 

importances = pd.Series(best_rf.feature_importances_, index=FEATURES).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
# Jo features 5% se zyada important hain unhe dark blue karo — baaki light blue
colors = ['#1f5e9e' if v > 0.05 else '#82aed4' for v in importances.values]
bars   = plt.barh(importances.index[::-1], importances.values[::-1], color=colors[::-1])
plt.xlabel('Importance Score')
plt.title('Feature Importance — Tuned Random Forest', fontweight='bold')
for bar, val in zip(bars, importances.values[::-1]):
    plt.text(val + 0.003, bar.get_y() + bar.get_height() / 2,
             f'{val:.3f}', va='center', fontsize=9)
plt.tight_layout()
plt.show()

# Printing Top 5 features 
print('\nTop 5 Features :')
print(importances.head())
```

# Time series Forecasting using ARIMA

```python
ts = (
    df.groupby(['Year', 'Month'])['Estimated_Deliveries']
      .sum()
      .reset_index()
      .sort_values(['Year', 'Month'])  #orderwise time sorting
)
ts['Date'] = pd.to_datetime(ts[['Year','Month']].assign(day=1))
ts = ts.set_index('Date')

#checkin anamolies using time series
plt.figure(figsize=(14, 4))
plt.plot(ts.index, ts['Estimated_Deliveries'], color='steelblue', linewidth=1.8)
plt.fill_between(ts.index, ts['Estimated_Deliveries'], alpha=0.12, color='steelblue')
plt.title('Monthly Global Tesla Deliveries (2015–2025)', fontweight='bold')
plt.ylabel('Deliveries')
plt.xlabel('Date')
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1000:.0f}K'))
plt.tight_layout()
plt.show()

print('Time series length:', len(ts), 'months')
```

```python
# ADF Test
# Stationary test — mean aur variance time ke saath change nahi karte
# ADF test se check karte hain  null hypothesis hai series non-stationary hai

ts_vals    = ts['Estimated_Deliveries'].values.astype(float)
adf_result = adfuller(ts_vals)

print('Augmented Dickey-Fuller Test')
print(f'ADF Statistic : {adf_result[0]:.4f}')
print(f'p-value       : {adf_result[1]:.6f}')
print('Critical Values:')
for key, val in adf_result[4].items():
    print(f'   {key}: {val:.4f}')

# p-value < 0.05  H0 reject — series stationary hai
# p-value >= 0.05  differencing zaroori hai — use d=1 ya d=2 in ARIMA  
if adf_result[1] < 0.05:
    print('\n Series STATIONARY - can use ARIMA')
else:
    print('\n Series NON-STATIONARY')
```

```python
# ACF aur PACF plots se ARIMA ke p aur q parameters decide karte hain
# ACF — q MA order decide karne mein help karta hai — kitne lags pe significant correlation hai
# PACF — p AR order decide karne mein help karta hai

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# ACF plot 
plot_acf(ts_vals, lags=30, ax=axes[0], color='steelblue', zero=False)
axes[0].set_title('Autocorrelation Function (ACF) — q decide karne ke liye', fontweight='bold')

# PACF plot 
plot_pacf(ts_vals, lags=30, ax=axes[1], color='tomato', method='ywm', zero=False)
axes[1].set_title('Partial Autocorrelation Function (PACF) — p decide karne ke liye', fontweight='bold')

plt.tight_layout()
plt.show()
# by seeing this plots we take p=2, d=1, q=2 
```

# ARIMA model fitting


```python
arima_model = ARIMA(ts_vals, order=(2, 1, 2))
arima_fit   = arima_model.fit()

print(arima_fit.summary())
```

```python
# 12 MONTHS FORECASTING
FORECAST_STEPS = 12

fc_obj  = arima_fit.get_forecast(steps=FORECAST_STEPS)
fc_mean = fc_obj.predicted_mean          # Point forecast — expected value
fc_ci   = fc_obj.conf_int()             # 95% confidence interval — uncertainty range

#Generating future dates for forecast
last_date = ts.index[-1]
fc_dates  = pd.date_range(start=last_date + pd.DateOffset(months=1),
                          periods=FORECAST_STEPS, freq='MS')

fc_df = pd.DataFrame({
    'Date'       : fc_dates,
    'Forecast'   : fc_mean.round(0).astype(int),      # Expected deliveries
    'Lower_95CI' : fc_ci[:, 0].round(0).astype(int),  # Worst case 95% CI
    'Upper_95CI' : fc_ci[:, 1].round(0).astype(int),  # Best case 95% CI
})
print('Next 12 Months Forecast ')
print(fc_df.to_string(index=False))
```

```python
#Virtual Representation
last24      = ts.index[-24:]                             # Last 2 saal ke actual dates
last24_vals = ts['Estimated_Deliveries'].values[-24:]    # Unki actual delivery values

plt.figure(figsize=(14, 5))

# Actual data — solid blue line
plt.plot(last24, last24_vals, color='steelblue', linewidth=2.2,
         label='Actual (last 24 months)', marker='o', markersize=3)

# Forecast — dashed orange line
plt.plot(fc_dates, fc_mean, color='tomato', linewidth=2.2,
         linestyle='--', label='Forecast', marker='s', markersize=4)

# Confidence interval — shaded area (jitna wider, utni uncertainty)
plt.fill_between(fc_dates, fc_ci[:, 0], fc_ci[:, 1],
                 color='tomato', alpha=0.15, label='95% Confidence Interval')

#  forecast  dotted line
plt.axvline(x=last24[-1], color='gray', linestyle=':', linewidth=1.5)
plt.text(last24[-1], plt.ylim()[1] * 0.97, ' Forecast\n start', color='gray', fontsize=9)

plt.title('ARIMA(2,1,2) — Monthly Delivery Forecast', fontweight='bold', fontsize=13)
plt.xlabel('Date')
plt.ylabel('Estimated Deliveries')
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1000:.0f}K'))
plt.legend()
plt.tight_layout()
plt.show()
```
